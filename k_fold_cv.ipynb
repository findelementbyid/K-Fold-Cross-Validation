```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('./HW2data.csv')
```

```python
from sklearn.preprocessing import LabelEncoder

# Process columns, apply LabelEncoder to categorical features
col = ['workclass', 'education', 'marital_status', 'occupation', 'relationship', 'race', 'sex', 'native_country', 'income']

for i in col:
    lbl = LabelEncoder()
    lbl.fit(list(data[i].values)) 
    data[i] = lbl.transform(list(data[i].values))
data
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
```

```python
def K_fold_CV(k, data):
    subsetSize = int(len(data) / k)
    accuracy = 0
    base = 0
    lim = subsetSize
    
    for i in range(k):
        
        test = data[base : lim]
        train = data[lim: ]
        temp = data[ : base]
        train = pd.concat((train, temp)).reset_index(drop = True)
        
        y_train = train['income']
        y_test = test['income']
        train.drop(['income'], axis=1, inplace = True)
        test.drop(['income'], axis=1, inplace = True)
        
        base = lim
        lim = base + subsetSize
        
        rd = RandomForestClassifier(n_estimators = 1000)
        rd.fit(train, y_train)
        y_pred = rd.predict(test)
        accuracy += metrics.accuracy_score(y_test, y_pred)

    return (accuracy / k)
```

```python
acc = K_fold_CV(10, data)
print(acc)
```
